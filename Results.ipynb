```python
mapping = {'mobilenet_full': 2225153, 'mobilenet_1_0_2': 1147265, 'mobilenet_0_75_0_3': 761545,
      'mobilenet_0_5_0_5': 534305, 'mobilenet_0_25_1': 237593
     }
```

```python
mura_df = pd.read_csv("imagenet_pretrained/mura_mobilenet.csv")
mura_df = mura_df.groupby(['model', 'dataset'])['score'].mean().reset_index()
mura_df['params'] = mura_df['model'].apply(lambda x: mapping[x])

chexpert_df = pd.read_csv("imagenet_pretrained/chexpert_mobilenet.csv")
chexpert_df = chexpert_df.groupby(['model', 'dataset'])['score'].mean().reset_index()
chexpert_df['params'] = chexpert_df['model'].apply(lambda x: mapping[x])

rsna_df = pd.read_csv("imagenet_pretrained/rsna_mobilenet.csv")
rsna_df = rsna_df.groupby(['model', 'dataset'])['score'].mean().reset_index()
rsna_df['params'] = rsna_df['model'].apply(lambda x: mapping[x])
```

```python
plt.plot(mura_df['params']/1e6, mura_df['score'], 'go--')
plt.plot(chexpert_df['params']/1e6, chexpert_df['score'], 'ro--')
plt.plot(rsna_df['params']/1e6, rsna_df['score'], 'bo--')
plt.legend(['Mura', 'Chexpert', 'RSNA'], loc='center right')
plt.xlabel("Parameters in Millions")
plt.ylabel("AUC")
plt.title("Transfer Learning")
plt.show()
```
